####  Chi-Square Test of Independence - Categorical Features

Chi-Square Test of independence is used to see if there is a significant association between categorical variables (Slavio, 2019). And it can be calculated with the following statistical formula:

![image](https://miro.medium.com/max/578/1*rGtwK5y2bxqYuhTR9nO2RA.webp)

In the formula, observed is the observed count for each category and expected is the expected count based on the population distribution for the corresponding category. 

This test aims to know the relation between the features and the target variable Churn=YES.
The test will determine if there is a significant difference in the proportion of Churns between the groups in the categorical variable.

The first value is the Chi-square value, followed by the p-value , then comes the degrees of freedom, and lastly, it outputs the expected frequencies as an array. 

First is created a contingency table with te variable "TechSupport"

```python
churnsample.groupby(['TechSupport', 'Churn'])['Churn'].count()
```

According with Roeschi (2020), only 20 percent of the above values can be less than 5 in this case, 100% of the values are bigger or = to 100%. As 83 was my minimum, is recommended to use  a sample of 1000 in order to not apply Yates correction.

It will be applied Pearson’s chi-squared test to examine whether these two variables, “TechSpport” and “Curn”, are statistically linked (= alternative hypothesis) or statistically independent (= null hypothesis). In this case, α-level at 0.05.

All codes gather from Roeschi (2020) report.

```python
# create contingency table with floats to avoid datatype issues with pd.DataFrame.at 
ar=np.array([[196, 48, 12],[280, 249, 215]])    
churnsample = pd.DataFrame(ar, columns=["No", "Yes", "No internet service"])
churnsample.index=["Churn","Not Churn" ] 
churnsample
```

Next it will be added the summ of each variable:

```python
df2=churnsample.copy() # create contingency table with the marginal totals and the grand total. 
df2.loc['Column_Total']= df2.sum(numeric_only=True, axis=0)
df2.loc[:,'Row_Total'] = df2.sum(numeric_only=True, axis=1)
df2
```

After is necessary to determine the expected values:

```python
n=df2.at["Column_Total", "Row_Total"]  # grand total 

exp=df2.copy()               # create dataframe with expected counts
for x in exp.index[0:-1]:
    for y in exp.columns[0:-1]:
        # round expected values to 6 decimal places to get the maximum available precision:
        v= (((df2.at[x, "Row_Total"]) * (df2.at["Column_Total", y])   )   /n ).round(6) 
        exp.at[x,y]=float(v)

exp = exp.iloc[[0, 1], [0,1,2]]
exp
```

None of the expected counts is <5.Below, it will be calculated the test-statistic:

```python
tstat = np.sum(((churnsample-exp)**2/exp).values) # calculate chi-squared test statistic
tstat
```

After, it will be obtained the degrees of freedom:

```python
dof = (len(churnsample.columns)-1)*(len(churnsample.index)-1) # determine degrees of freedom 
dof
```

Now, it will be calculated the corresponding p-value by using the cumulative distribution function given the degrees of freedom:

```python
pval=2-chi2.cdf(tstat, dof) # subtract the cumulative distribution function from 1
pval
```

Finally, verify the result by comparing it to Python’s built-in function scipy.stats.chi2_contingency by using“correction=False”:

```python
from scipy.stats import chi2_contingency # import Scipy's built-in function

tstat_scipy,pval_scipy,ddof_scipy,exp_scipy=chi2_contingency(churnsample, correction=False) # "correction=False" means no Yates' correction is used! 
print("Chi-squared test statistic without Yates correction (Scipy): " + str(tstat_scipy))
print("P-value without Yates correction (Scipy): " + str(pval_scipy))
```

From the two results, its can be seen Chi squared tests with idenical results, only with a small difference on the pvalue result. As it can be seen, p-value is more than α-level of 0.05. Hence, it cannot reject the null hypothesis.

```python
# this function creates will automaticall generate the decision for the chi square test
def chi_square_test(data, feature):
    df = pd.crosstab(data[feature],data['Churn'])
    stat, p, dof, expected = chi2_contingency(df)
    print('='*60)
    print('Chi-Square Test for {}'.format(feature))
    print('='*60)
    print('degrees of freedom = %d' % dof)
    print('-'*60)
    # interpret test-statistic
    prob = 0.95
    critical = chi2.ppf(prob, dof)
    print('Results:')
    print(' ')
    print('probability = %.3f, critical = %.3f, stat = %.3f' % (prob, critical, stat))
    print('-'*60)
    print('Decision based on Chi-Square Statistics:')
    print(' ')
    if abs(stat) >= critical:
        print('The Churns are dependent on {} : (reject H0)'.format(feature))
    else:
        print('The Churns are independent on {} : (fail to reject H0)'.format(feature))
    # interpret p-value
    alpha = 1.0 - prob
    print('-'*60)
    print('Decision based on p-value:')
    print(' ')
    print('alpha = %.2f, p = %.3f' % (alpha, p))
    if p <= alpha:
        print('The Churns are dependent on {} : (reject H0)'.format(feature))
    else:
        print('The Churns are independent on {} : (fail to reject H0)'.format(feature))
    print('_'*60)
    print(' '*60)
    print('_'*60)
    print('_'*60)
```

#### ANOVA Test of Difference - Numerical Features

ANOVA is a statistical test to analyse if there is a relevant difference among two or more categorical groups by testing its differences of means using variance (Rowntree, 2019).
This test will analyse if there is difference between the means of Churn=YES and Churn=NO in numerical variables Tenure, Total Charges and Monthly Charges.

* Hypotheses:

H0 = assumes no statistically significant difference in means between the Churn groups.\
H1 = assumes there is a significant difference between the group.

Samples are already in the data, which contains 50 samples.

The alpha statistics is set to 0.05%.
If the p-value is lesser than alpha, then we will reject the null hypothesis and accept the alternative that there is a significant difference between the groups.
